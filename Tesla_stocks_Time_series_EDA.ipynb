```python
tesla_df.head()
```

```python
tesla_df.resample(rule='A').min()
tesla_df.resample(rule='A').max()
```

```python
tesla_df.resample(rule='A').max()["Open"].plot()
```

```python
tesla_df.resample(rule="QS").max()["Open"].plot()
```

```python
tesla_df.resample(rule="BA").max()
```

```python
tesla_df.resample(rule="BQS").max()
```

```python
tesla_df["High"].resample(rule="BA").mean().plot(kind="bar")
```

```python
tesla_df["Open"].resample(rule="M").mean().plot(kind="bar",figsize=(15,6))
```
